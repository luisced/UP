```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
# abrimos una imagen con cv

img = cv.imread(r"C:\Users\emili\OneDrive\Escritorio\UP\TERCER_SEMESTRE\Procesamiento_digital_de_imagenes\Imagenes\messi5.jpg", )[:,:,::-1]

#mostramos la imagen

plt.imshow(img)
```

## Vamos a separar la imagen con sus diferentes canales de color

```python
# Existen dos maneras de separarlas
# R G B

canal_R = img[: ,: ,0]

canal_G = img[:, :, 1]

canal_B = img[:, :, 2]

plt.imshow(canal_R, cmap="gray") # la mostramos en escala de grises, si vemos muchos pixeles negros, significa la ausencia del rojo, los más blancos significa
#que hay más rojos
```

```python
# Vamos a imprimir las tres

plt.title("Canal R")
plt.imshow(canal_R, cmap="gray")

canal_R.shape # ya no nos dice sobre la tercera dimension sobre el color de la matriz
```

```python
plt.title("Canal G")
plt.imshow(canal_G, cmap="gray")

canal_G.shape
```

```python
plt.title("Canal B")
plt.imshow(canal_B, cmap="gray")

canal_B.shape
```

>### Podemos generar una sola imagen con todos los plots

```python
plt.subplot(131) # 13 significa 1 fila 3 columnas posteriormente metemos secuancialmente los elementos, por eso hay un 1 despues del 13
plt.title("Canal R")
plt.imshow(canal_R, cmap="gray")

plt.subplot(132) 
plt.title("Canal G")
plt.imshow(canal_G, cmap="gray")

plt.subplot(133)
plt.title("Canal B")
plt.imshow(canal_B, cmap="gray")
```

```python
# podemos mandar las tres instrucciones en una sola linea de la siguiente manera

plt.subplot(131);plt.title("Canal R");plt.imshow(canal_R, cmap="gray")
plt.subplot(132);plt.title("Canal G");plt.imshow(canal_G, cmap="gray")
plt.subplot(133);plt.title("Canal B");plt.imshow(canal_B, cmap="gray")
```
