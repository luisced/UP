```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

```python
# Cargar la imagen
imagen = cv.imread("./IMG_7863_Original.JPG", 0)
```

```python
# Mostrar Imagen en gris
plt.imshow(imagen, "gray")
```

```python
histograma = cv.calcHist([imagen], [0], None, [256], (0,255))
```

```python
#plt
plt.subplot(1, 2, 1)
plt.title("Chicky")
plt.imshow(imagen, "gray")
plt.axis("off")


plt.subplot(1, 2, 2)
plt.title("Histograma")
plt.plot(histograma)
#caption 
plt.xlabel("Cantidad de pixeles entre 0 y 255")

plt.tight_layout()
```

## Histograma a Color

```python
img = cv.imread("./IMG_7863_Original.JPG",)
#rgb
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
 #show image rgb

plt.imshow(img)
plt.axis("off")
```

```python
r, g, b = cv.split(img)
```

```python
histograma_r = cv.calcHist([r], [0], None, [256],(0, 255))
histograma_g = cv.calcHist([g], [0], None, [256],(0, 255))
histograma_b = cv.calcHist([b], [0], None, [256],(0, 255))

plt.subplot(1, 3, 1)
plt.plot(histograma_r, "r")
plt.title("R")

plt.subplot(1, 3, 2)
plt.plot(histograma_g, "g")
plt.title("G")

plt.subplot(1, 3, 3)
plt.plot(histograma_b, "b")
plt.title("B")

plt.tight_layout()

```
