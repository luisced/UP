```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Cargar la imagen
imagen = plt.imread("./chicky_512.png")
```

```python
# Definir el factor de corrección Gamma
t1 = 0.8
t2 = 2.0
```

```python
# Aplicar la corrección Gamma
imagen_corregida_1 = np.power(imagen % 256.0, t1)
imagen_corregida_2 = np.power(imagen % 256.0, t2)
```

```python
# Mostrar las imágenes corregidas utilizando Matplotlib
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.title("Imagen Original")
plt.imshow(imagen, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title(f"Corrección Gamma ({t1})")
plt.imshow(imagen_corregida_1, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title(f"Corrección Gamma ({t2})")
plt.imshow(imagen_corregida_2, cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()
```
