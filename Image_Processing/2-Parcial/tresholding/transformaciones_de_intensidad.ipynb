# Transformaciones de identidad y filtrado espacial

## Introducción

El dominio espacial opera directamente sobre los pixeles de una imagen a diferencia por ejemplo del dominio de la frecuencia en el que las operaciones se realizan dobre la transformada de furier

### Ecuación #1

$ g(x,y) = T[f(x,y)] $

donde $ f(x,Y) $ es la imagen original, $g(x,y)$ es el operador de salida

Vamos a usar la vecindad más pequeña que es la vecindad de 1, es decir el pixel único $p$

### Funciones de transformacion de identidad o función treshholding

Todos los pixeles que pasen del umbral (128) se van a mandar a 255 y los que no lo pasen se van a pasar a 0, por lo que vamos a tener como resultado una imagen binarizada, o una imagen en blanco y negro

### Paso #1 leemos la imagen en escala de grises

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

image = cv.imread("chicky_512.png",0)

plt.imshow(image, cmap="gray")
```

# Paso #2 creamos una matriz con las mismas dimensiones

```python
matriz = np.zeros((image.shape[0], image.shape[1]), dtype = np.uint8)

plt.imshow(matriz, cmap= "gray")
```

# Paso #3 aplicamos la condicion y asignamos los nuevos valores en la matriz

```python
for i in range(0, image.shape[0], 1):
    for l in range(0, image.shape[1], 1):
        if(image[i][l] >= 125):
            matriz[i][l] = 255
        else:
            matriz[i][l] = 0

plt.imshow(matriz, cmap="gray")
```

### Llamamos a esta tecnica umbralizacion global en donde ponemos un unmbral en toda la imagen

Ahora, existe un metodo para realizar este metodo de umbralizacion que lo hace más facilmente y es la siguiente:

```python
cv.threshold(image, 128, 1, cv.THRESH_BINARY)
```

Tenemos que plotear el segundo elemento de la tupla, por que esa es la imagen

```python
cv.threshold(image, 128, 1, cv.THRESH_BINARY)[1]
```

```python
#ploteamos

plt.imshow(cv.threshold(image, 128, 1, cv.THRESH_BINARY)[1], cmap="gray")
```

### Podemos hacerlo de la manera invertida con la umbralizacion

```python
plt.imshow(cv.threshold(image, 128, 1, cv.THRESH_BINARY_INV)[1], cmap="gray")
```

### podemos usar una funcion para corroborar los valores de nuestra imagen y si es binaria

```python
np.unique(matriz)
```

# NEGATIVOS


```python
imagen_aux = np.zeros((image.shape[0], image.shape[1]), dtype = np.uint8)

plt.imshow(imagen_aux, cmap= "gray")
```

```python
for i in range(0, image.shape[0], 1):
    for l in range(0, image.shape[1], 1):
        imagen_aux[i][l] = 255 - image[i][l]

plt.imshow(imagen_aux, cmap="gray")
```

```python
plt.imshow(255 - imagen_aux, cmap="gray")
```
